## Task 2: X API as LangChain Tools

Instead of relying on a community-built MCP server for X, we'll call the **X API v2** directly and wrap our functions with the `@tool` decorator. This makes them available to our LangGraph agent as callable tools.

This is a key architectural decision: **not everything needs to be an MCP server**. Wrapping a simple API call as a `@tool` is often simpler and more transparent.

**Documentation:**
- [LangChain Tools Conceptual Guide](https://python.langchain.com/docs/concepts/tools/)
- [X API v2 Documentation](https://developer.x.com/en/docs/x-api)

````python
import requests
import json
from langchain_core.tools import tool

BEARER_TOKEN = os.environ.get("X_BEARER_TOKEN")


@tool
def search_recent_posts(query: str, max_results: int = 20) -> str:
    """Search recent X/Twitter posts using the v2 API.
    Returns posts from the last 7 days matching the query.
    Use this for keyword searches, hashtag searches, or general topic searches.

    Args:
        query: The search query (e.g., 'Stone Ridge', '#bitcoin', 'reinsurance market')
        max_results: Number of results to return (10-100, default 20)
    """
    url = "https://api.x.com/2/tweets/search/recent"
    headers = {"Authorization": f"Bearer {BEARER_TOKEN}"}
    params = {
        "query": query,
        "max_results": min(max(max_results, 10), 100),
        "tweet.fields": "created_at,public_metrics,author_id,text",
        "expansions": "author_id",
        "user.fields": "name,username",
    }

    response = requests.get(url, headers=headers, params=params)
    response.raise_for_status()
    data = response.json()

    tweets = data.get("data", [])
    if not tweets:
        return "No posts found for this query."

    result_lines = [f"Found {len(tweets)} posts:\n"]
    for t in tweets:
        metrics = t.get("public_metrics", {})
        result_lines.append(
            f"[{t.get('created_at', 'unknown')[:10]}] "
            f"{t['text'][:200]}\n"
            f"  Likes: {metrics.get('like_count', 0)} | "
            f"Retweets: {metrics.get('retweet_count', 0)}"
        )
    return "\n\n".join(result_lines)


@tool
def get_user_posts(username: str, max_results: int = 20) -> str:
    """Get recent original posts (no retweets) from a specific X/Twitter user.
    Use this when you want to see what a specific account has been posting.

    Args:
        username: The X/Twitter handle without the @ sign (e.g., 'StoneRidgeAM')
        max_results: Number of results to return (10-100, default 20)
    """
    query = f"from:{username} -is:retweet"
    return search_recent_posts.invoke({"query": query, "max_results": max_results})


x_api_tools = [search_recent_posts, get_user_posts]
print(f"Created {len(x_api_tools)} X API tools: {[t.name for t in x_api_tools]}")
````

````python
from typing import Annotated, Literal
from typing_extensions import TypedDict

from langgraph.graph import StateGraph, START, END
from langgraph.graph.message import add_messages
from langgraph.prebuilt import ToolNode
from langgraph.checkpoint.memory import MemorySaver
from langchain_core.messages import HumanMessage, AIMessage, SystemMessage

# Combine all tools: X API tools + GitHub MCP tools
all_tools = x_api_tools + github_mcp_tools
print(f"Total tools available to agent: {len(all_tools)}")
print(f"  X API tools: {[t.name for t in x_api_tools]}")
print(f"  GitHub MCP tools: {[t.name for t in github_mcp_tools]}")
````

````python
# Step 1: Define the Agent State
class AgentState(TypedDict):
    messages: Annotated[list, add_messages]


# Step 2: Define the system prompt
SYSTEM_PROMPT = """You are an AI investment research assistant that can search X/Twitter posts for \
market sentiment and manage GitHub repositories.

You have two categories of tools:
1. X API tools: search_recent_posts, get_user_posts \u2014 for searching and retrieving X/Twitter posts \
about investments, market sentiment, Stone Ridge, bitcoin, reinsurance, etc.
2. GitHub MCP tools: for creating repos, committing files, creating branches, opening PRs, etc.

When asked to analyze market sentiment, retrieve posts first using the X API tools, then provide \
a structured markdown summary with: Overview, Key Themes, Notable Posts, and Summary Statistics.

When asked to perform GitHub operations, use the appropriate GitHub MCP tool.
Always use the available tools when appropriate. Be concise in your responses."""


# Step 3: Bind tools to the LLM
llm_with_tools = llm.bind_tools(all_tools)


# Step 4: Define the agent node
def agent_node(state: AgentState):
    """The agent node \u2014 calls the LLM with the current conversation and available tools."""
    messages = [SystemMessage(content=SYSTEM_PROMPT)] + state["messages"]
    response = llm_with_tools.invoke(messages)
    return {"messages": [response]}


# Step 5: Define the tool node
tool_node = ToolNode(all_tools, handle_tool_errors=True)


# Step 6: Define routing logic
def should_continue(state: AgentState) -> Literal["tools", "end"]:
    """Determine whether to call tools or end the conversation."""
    last_message = state["messages"][-1]
    if hasattr(last_message, "tool_calls") and last_message.tool_calls:
        return "tools"
    return "end"


# Step 7: Build the graph
workflow = StateGraph(AgentState)

workflow.add_node("agent", agent_node)
workflow.add_node("tools", tool_node)

workflow.add_edge(START, "agent")
workflow.add_conditional_edges("agent", should_continue, {"tools": "tools", "end": END})
workflow.add_edge("tools", "agent")

# Compile with MemorySaver for short-term memory across turns
checkpointer = MemorySaver()
agent = workflow.compile(checkpointer=checkpointer)

print("Agent compiled with memory and tools!")
````

````python
from IPython.display import Image, display

display(Image(agent.get_graph().draw_mermaid_png()))
````

````python
# Use a consistent thread_id so the agent remembers context across all tasks
config = {"configurable": {"thread_id": "mcp-investment-workflow-1"}}


async def ask_agent(user_message: str) -> str:
    """Send a message to the agent and return its final response."""
    response = await agent.ainvoke(
        {"messages": [HumanMessage(content=user_message)]},
        config,
    )
    return response["messages"][-1].content
````

### ❓ Question #1

How could X/Twitter sentiment analysis complement traditional investment research methods like reading investor letters?

##### Answer

X/Twitter sentiment analysis complements traditional investment research in several key ways:

1. **Real-time signal vs. periodic reporting**: Investor letters are published quarterly or annually and reflect a firm's curated, backward-looking perspective. X/Twitter provides a continuous, real-time stream of market reactions — capturing sentiment shifts within hours of news events (e.g., rate decisions, catastrophe events, regulatory changes). This fills the information gap between formal publications.

2. **Breadth of market perspectives**: An investor letter represents a single firm's view. X aggregates opinions from thousands of market participants — analysts, fund managers, retail investors, journalists — giving a broader "wisdom of the crowd" signal. This helps identify consensus views vs. contrarian positions that a single letter may not surface.

3. **Narrative and theme detection**: X posts often surface emerging themes before they appear in formal research. For example, shifting sentiment around bitcoin institutional adoption or reinsurance pricing hardening may appear on X weeks before it's formalized in analyst reports. Tracking trending topics and hashtags can serve as an early-warning system for thematic shifts.

4. **Validation and calibration**: X sentiment can validate (or challenge) claims in investor letters. If Stone Ridge's letter is bullish on reinsurance but X sentiment from cat bond traders is bearish, that divergence is valuable signal. It helps calibrate how much weight to place on a firm's stated outlook.

5. **Limitations to be aware of**: X data is noisy, subject to bot manipulation, biased toward retail sentiment, and lacks the rigor of formal research. It should be treated as a complementary input — a sentiment overlay on top of fundamental analysis — not a replacement for deep research.

````python
"""
Activity #1: Create a get_user_profile tool using X API v2 user lookup endpoint.
"""

import requests
import os
from langchain_core.tools import tool

BEARER_TOKEN = os.environ.get("X_BEARER_TOKEN")


@tool
def get_user_profile(username: str) -> str:
    """Get public profile information for a specific X/Twitter user.
    Returns the user's display name, bio, follower/following counts, and account creation date.

    Args:
        username: The X/Twitter handle without the @ sign (e.g., 'elaborateidea', 'StoneRidgeAM')
    """
    url = f"https://api.x.com/2/users/by/username/{username}"
    headers = {"Authorization": f"Bearer {BEARER_TOKEN}"}
    params = {
        "user.fields": "description,public_metrics,created_at,verified,location,url",
    }

    try:
        response = requests.get(url, headers=headers, params=params, timeout=15)
        response.raise_for_status()
    except requests.RequestException as exc:
        return f"X API request failed: {exc}"

    data = response.json()
    user = data.get("data", {})

    if not user:
        errors = data.get("errors", [])
        if errors:
            return f"User not found: {errors[0].get('detail', 'Unknown error')}"
        return f"User '@{username}' not found."

    metrics = user.get("public_metrics", {})
    profile = (
        f"Profile: @{user.get('username', username)}\n"
        f"Name: {user.get('name', 'N/A')}\n"
        f"Bio: {user.get('description', 'No bio')}\n"
        f"Location: {user.get('location', 'Not specified')}\n"
        f"Followers: {metrics.get('followers_count', 0):,}\n"
        f"Following: {metrics.get('following_count', 0):,}\n"
        f"Posts: {metrics.get('tweet_count', 0):,}\n"
        f"Joined: {user.get('created_at', 'Unknown')[:10]}\n"
        f"URL: {user.get('url', 'None')}"
    )
    return profile


# --- Test the tool directly ---
print("=== Testing get_user_profile ===\n")
result = get_user_profile.invoke({"username": "anthropaborian"})
print(result)

# --- Rebuild the agent with the new tool ---
print("\n\n=== Rebuilding agent with get_user_profile ===\n")

# Add the new tool to the tool set
updated_x_tools = [search_recent_posts, get_user_posts, get_user_profile]
updated_all_tools = updated_x_tools + github_mcp_tools

print(f"Updated tool set: {[t.name for t in updated_x_tools]}")
print(f"Total tools: {len(updated_all_tools)}")

# Rebuild the agent with the updated tools
from langgraph.graph import StateGraph, START, END
from langgraph.prebuilt import ToolNode
from langgraph.checkpoint.memory import MemorySaver
from langchain_core.messages import HumanMessage, SystemMessage

llm_updated = llm.bind_tools(updated_all_tools)


def updated_agent_node(state):
    messages = [SystemMessage(content=SYSTEM_PROMPT)] + state["messages"]
    return {"messages": [llm_updated.invoke(messages)]}


updated_workflow = StateGraph(AgentState)
updated_workflow.add_node("agent", updated_agent_node)
updated_workflow.add_node("tools", ToolNode(updated_all_tools, handle_tool_errors=True))
updated_workflow.add_edge(START, "agent")
updated_workflow.add_conditional_edges("agent", should_continue, {"tools": "tools", "end": END})
updated_workflow.add_edge("tools", "agent")

updated_agent = updated_workflow.compile(checkpointer=MemorySaver())
print("Updated agent compiled!")

# Test: ask about a financial thought leader
config_new = {"configurable": {"thread_id": "profile-test-1"}}
result = await updated_agent.ainvoke(
    {"messages": [HumanMessage(
        content="Look up the X profile for @unusual_whales and get their recent posts. "
        "Summarize who they are and what they've been posting about."
    )]},
    config_new,
)
print("\n=== Agent Response ===\n")
print(result["messages"][-1].content[:1500])
````

### ❓ Question #2

Compare using GitHub MCP tools (through a LangGraph agent) to traditional `git` commands. What felt easier? What felt harder or less transparent?

##### Answer

**What felt easier with GitHub MCP tools:**

- **Natural language orchestration**: Instead of remembering exact git syntax (`git checkout -b`, `git push -u origin`, `gh pr create --body "..."`), you describe the intent ("create a branch called add-metadata and commit the file") and the agent handles the mechanics. This dramatically lowers the barrier for complex multi-step workflows.
- **End-to-end workflows in one conversation**: Creating a repo, committing files, branching, and opening a PR all happened in a single agent session with memory. With traditional git, each step is a separate command that you'd need to sequence manually.
- **No local repo state needed**: The MCP tools operate directly on the GitHub remote API — no need to clone, manage a working directory, resolve merge states, or push. This is especially convenient for one-off operations.

**What felt harder or less transparent:**

- **Lack of visibility into what happened**: With `git log`, `git diff`, and `git status`, you can inspect exactly what changed at every step. With MCP tools through an agent, you're trusting the agent's summary of what it did. If something goes wrong, debugging is harder — you can't see the intermediate API calls or responses without checking LangSmith traces.
- **Less control over exact operations**: Git gives you precise control (amend, interactive rebase, cherry-pick, stash). MCP tools expose a subset of GitHub's API — you can't do everything git can. For example, resolving merge conflicts or rebasing interactively is not feasible through MCP tools.
- **Error handling is opaque**: If a git command fails locally, the error message is immediate and specific. If an MCP tool call fails inside the agent, the agent may retry, hallucinate a success, or give a vague error message. You need LangSmith tracing to understand what actually happened.
- **Latency**: Each MCP tool call involves an API round-trip through the agent, which is significantly slower than local git operations.

### ❓ Question #3

You used MCP for GitHub but wrapped the X API as a `@tool` directly. What are the tradeoffs of consuming an API through an MCP server versus wrapping it as a LangChain tool? When would each approach make more sense?

##### Answer

**MCP server approach (used for GitHub):**

| Pros | Cons |
|------|------|
| Reusable across any MCP-compatible client (Claude Desktop, LangGraph, etc.) | Extra process/transport overhead (subprocess for stdio, HTTP for remote) |
| Decoupled — server team and agent team can iterate independently | More infrastructure to manage (separate server process, health checks) |
| Standardized discovery — client auto-discovers available tools | Debugging requires tracing through the MCP protocol layer |
| Official servers exist for popular services (GitHub, Slack, etc.) | Schema changes require server redeployment |

**Direct `@tool` wrapper approach (used for X API):**

| Pros | Cons |
|------|------|
| Zero overhead — just a Python function call | Tightly coupled to LangChain/LangGraph framework |
| Full control over implementation, error handling, retries | Not reusable outside of LangChain ecosystem |
| Easy to debug (set breakpoints, inspect variables) | Every consuming agent needs to redefine the tool |
| Fast iteration during development | No standardized discovery — tools are hardcoded |

**When each makes more sense:**

- **Use MCP when**: The tool set is mature and will be consumed by multiple clients/teams. GitHub is a perfect example — the official MCP server covers dozens of operations, is maintained by GitHub, and any MCP client can use it. Also prefer MCP when you want to separate tool infrastructure from agent logic (e.g., a platform team maintains MCP servers, application teams build agents).

- **Use `@tool` when**: You're wrapping a simple API with 1-3 endpoints, iterating quickly during development, need custom logic that doesn't fit the MCP tool model (e.g., chaining multiple API calls in a single tool), or performance matters (avoiding subprocess/transport overhead). The X API wrapper was a good fit because we only needed two functions and wanted tight control over query construction and response formatting.

- **Hybrid approach** (as used in this notebook): Use MCP for well-supported, multi-operation services where official servers exist, and `@tool` for custom or lightweight integrations. This gives you the best of both worlds.
